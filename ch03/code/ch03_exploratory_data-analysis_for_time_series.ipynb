# Exploratory Data Analysis for Time Series

## Familiar Methods

### Plotting

```r
# Daily closing prices of 4 major EU stock indexes from 1991 to 1998
# Data set only includes business days
head(EuStockMarkets)
```

```r
# Plot the data
plot(EuStockMarkets)
```

```r
# Conventient functions of `ts` class
frequency(EuStockMarkets)
start(EuStockMarkets)
end(EuStockMarkets)
```

```r
# Window temporal section of the data
window(EuStockMarkets, start=1997, end=1998)
```

### Histograms

```r
hist(EuStockMarkets[,'SMI'], 30)
hist(diff(EuStockMarkets[,'SMI'],30))
```

### Scatter Plots

```r
plot(EuStockMarkets[,'SMI'], EuStockMarkets[,'DAX'])
plot(diff(EuStockMarkets[,'SMI']), diff(EuStockMarkets[,'DAX']))
```
