## Time Series-Specific Exploratory Data Analysis Methods

### Stationarity

#### Applying Window Functions

```r
# Rolling windows
# Calculate rolling average using the base R filter function
x <- rnorm(n=100, mean=0, sd=10) + 1:100
mn <- function(n) rep(1/n, n)

plot(x, type='l', lwd=1)
lines(filter(x, mn(5)), col=2, lwd=3, lty=2) # Use 5-point window
lines(filter(x, mn(50)), col=3, lwd=3, lty=3) # Use 50-point window
```

```r
# Rolling windows
# Use non linear transformation-based rolling windows
require(zoo)

f1 <- rollapply(zoo(x), 20, function(w) min(w), align='left', partial=TRUE)
f2 <- rollapply(zoo(x), 20, function(w) min(w), align='right', partial=TRUE)

plot(x, lwd=1, type='l')
lines(f1, col=2, lwd=3, lty=2)
lines(f2, col=3, lwd=3, lty=3)
```

```r
# Expanding windows
plot(x, type='l', lwd=1)
lines(cummax(x), col=2, lwd=3, lty=2) # max
lines(cumsum(x)/1:length(x), col=3, lwd=3, lty=3) # mean
```

### Autocorrelation

```r
# Autocorrelation function (ACF)
x <- 1:100
y <- sin(x*pi/3)
plot(y, type='b')
acf(y)
```

```r
# Autocorrelation function (ACF)
require(data.table)

cor(y, shift(y, 1), use='pairwise.complete.obs')
cor(y, shift(y, 2), use='pairwise.complete.obs')
```

```r
# Partial autocorrelation function (PACF)
y <- sin(x*pi/3)
plot(y[1:30], type='b')
pacf(y)
```

```r
# Noisy functions
y1 <- sin(x*pi/3)
plot(y1, type='b')
acf(y1)
pacf(y1)
```

```r
y2 <- sin(x*pi/10)
plot(y2, type='b')
acf(y2)
pacf(y2)
```

```r
y <- y1 + y2
plot(y, type='b')
acf(y)
pacf(y)
```

```r
noise1 <- rnorm(100, sd=0.05)
noise2 <- rnorm(100, sd=0.05)

y1 <- y1 + noise1
y2 <- y2 + noise2
y <- y1 + y2

plot(y1, type='b')
acf(y1)
pacf(y1)

plot(y2, type='b')
acf(y2)
pacf(y2)

plot(y, type='b')
acf(y)
pacf(y)
```

```r
# Nonstationary data
x <- 1:100
plot(x)
acf(x)
pacf(x)
```
